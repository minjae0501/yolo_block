```python
import cv2
import mediapipe as mp
from ultralytics import YOLO
import torch
```

```python
# YOLO 객체 감지 모델 초기화
# best_model = '/content/drive/MyDrive/Deep_learning_Team_PJ/hand_tracing/yolo_eraser/best.pt'
best_model = 'truck_best_04.pt'
# best_model = 'block_best_02.pt'
# best_model = 'truck_block_all_best_02.pt'
yolo_model = YOLO(best_model)

device = 0
cap = cv2.VideoCapture(device) # for Mac
# cap = cv2.VideoCapture(0) # for Windows

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    image = cv2.resize(frame, (640, 640))

    # YOLO 객체 감지
    box_results = yolo_model.predict(image, conf = 0.6)
    
    res_plot = box_results[0].plot()

    cv2.imshow('frame',res_plot)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
